Example of How Gaussian Graphs Work
===================

```r
getGaussianValue=function (x,mu, sigma, scaling){
    val=scaling*(1/ (sqrt(2*pi)*sigma ))*exp(-0.5*((x-mu)/sigma)^2 )
    return(val)
}
```

```r
X=seq(0,25,by=0.1)
```

```r
Y1=sapply(X,getGaussianValue,mu=15,sigma=1,scaling=1)
Y2=sapply(X,getGaussianValue,mu=14,sigma=1,scaling=0.5)
Y3=sapply(X,getGaussianValue,mu=5,sigma=1,scaling=1)
```

```r
plot(X,Y1,type='l',xlab="Sample Number",ylab="Expression Value")
lines(X,Y2,type='l',col='red')
lines(X,Y3,type='l',col='blue')
```

```r
OriginalMatrix=rbind(Y1,Y2,Y3)
```

```r
correlationMatrix=cor(t(OriginalMatrix))
```

```r
correlationMatrix
```

```r
abs(solve(correlationMatrix))
```

```r
abs(solve(correlationMatrix))>0.1
```

```r
#Adding batch effect
Y1=Y1+100
Y2=Y2+100
Y3=Y3+100
```

```r
newMatrix=rbind(Y1,Y2,Y3)
```

```r
newCorrelationMatrix=cov(t(newMatrix))
newCorrelationMatrix
```

```r
abs(solve(newCorrelationMatrix))/max(solve(newCorrelationMatrix))>0.05
```
